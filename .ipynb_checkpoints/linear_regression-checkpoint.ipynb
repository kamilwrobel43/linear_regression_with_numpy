# Linear Regression Function

```python
import numpy as np

def calculate_preds(X, a, b):
    return np.dot(X, a) + b

def MSE(y, y_pred):
    n = y_pred.shape[0]
    mse = (1/n) * np.sum((y - y_pred) ** 2)
    return mse

def linear_regression(X, y, iter=10000, alfa=0.001):  
    if X.ndim == 1:
        X=X.reshape(-1,1)
        m, n =X.shape
    else:
        m, n = X.shape
        
    a=np.zeros(n)
    b = 0  
    
    for i in range(iter):
        y_pred = calculate_preds(X, a, b)
        da = (-2/m) * np.dot(X.T, (y - y_pred)) 
        db = (-2/m) * np.sum(y - y_pred) 

        a = a - alfa * da
        b = b - alfa * db
    
    return a, b

a, b = linear_regression(X, y_noisy)
print(a, b)
```
